### Convert time into datetime

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat
from datetime import datetime, timedelta
import matplotlib.dates as mdates

# Load .mat file
data = loadmat('095034_20190707_1550_data.mat')

# Access variables
variable1 = data['thistemp']
variable2 = data['thistime']

# Convert thistime to a numpy array and flatten it
time_values = np.array(variable2).flatten()

# Convert MATLAB date numbers to datetime objects
def matlab_to_datetime(matlab_datenum):
    return datetime.fromordinal(int(matlab_datenum)) + timedelta(days=matlab_datenum%1) - timedelta(days=366)

time_objects = [matlab_to_datetime(matlab_datenum) for matlab_datenum in time_values]

# Plot the time series
plt.figure(figsize=(10, 6))
plt.plot(time_objects, variable1, label='Temperature')
plt.title('Temperature Time Series from Thermistor Array')
plt.xlabel('Time')
plt.ylabel('Temperature [°C]')

# Format x-axis ticks as hours:minutes
plt.gca().xaxis.set_major_locator(mdates.HourLocator(interval=8))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))

# Rotate x-axis tick labels
plt.xticks(rotation=45)

plt.legend()
plt.show()
```

# Make a map of vent sites and axial seamount

```python
import cartopy.crs as ccrs
import cartopy
import matplotlib.pyplot as plt
%matplotlib inline

from glob import glob

import matplotlib
import numpy as np

import cartopy.feature as cfeature
import cartopy.io.shapereader as shpreader

def load_bathymetry(zip_file_url):
    """Read zip file from Natural Earth containing bathymetry shapefiles"""
    # Download and extract shapefiles
    import io
    import zipfile

    import requests
    r = requests.get(zip_file_url)
    z = zipfile.ZipFile(io.BytesIO(r.content))
    z.extractall("ne_10m_bathymetry_all/")

    # Read shapefiles, sorted by depth
    shp_dict = {}
    files = glob('ne_10m_bathymetry_all/*.shp')
    assert len(files) > 0
    files.sort()
    depths = []
    for f in files:
        depth = '-' + f.split('_')[-1].split('.')[0]  # depth from file name
        depths.append(depth)
        bbox = (-115, 35, -140, 60)  # (x0, y0, x1, y1)
        nei = shpreader.Reader(f, bbox=bbox)
        shp_dict[depth] = nei
    depths = np.array(depths)[::-1]  # sort from surface to bottom
    return depths, shp_dict
    
if __name__ == "__main__":
    # Load data (14.8 MB file)
    depths_str, shp_dict = load_bathymetry(
        'https://naturalearth.s3.amazonaws.com/' +
        '10m_physical/ne_10m_bathymetry_all.zip')
    
     # Construct a discrete colormap with colors corresponding to each depth
    depths = depths_str.astype(int)
    N = len(depths)
    nudge = 0.01  # shift bin edge slightly to include data
    boundaries = [min(depths)] + sorted(depths+nudge)  # low to high
    norm = matplotlib.colors.BoundaryNorm(boundaries, N)
    blues_cm = matplotlib.colormaps['Blues_r'].resampled(N)
    colors_depths = blues_cm(norm(depths))

    # Set up plot
    fig = plt.figure()  # Add this line to create a figure
    extent = [-120, -135, 40, 50]
    #Axial Seamount: -130, 46.06
    #Endeavour: -129.1, 47.95
    #ASHES: 	-130.0167, 45.9333
    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.gridlines()
    ax.coastlines()
    ax.set_extent(extent)

    # Add state lines
    states = cfeature.NaturalEarthFeature(category='cultural',
                                        name='admin_1_states_provinces_lines',
                                        scale='50m',
                                        facecolor='none')
    ax.add_feature(states, edgecolor='black')

    # Add features (vent sites)
    ax_lon = -130
    ax_lat = 46.06
    ashes_lon = -130.013672
    ashes_lat = 45.933643

    font = {'family': 'Times New Roman', 'weight': 'normal'}

    ax.plot(ax_lon, ax_lat, marker="o", markersize=14, markeredgecolor="black", markerfacecolor="red", linestyle='None', label="Axial Seamount")
    ax.plot(ashes_lon, ashes_lat, marker="o", markersize=6, markeredgecolor="black", markerfacecolor="blue", linestyle='None',label="ASHES Vent Field = 1540(m)")

    ax.legend()
    plt.title('Map of Vent Field and Bathymetry', fontdict=font)

    # Iterate and plot feature for each depth level
    for i, depth_str in enumerate(depths_str):
        ax.add_geometries(shp_dict[depth_str].geometries(),
                          crs=ccrs.PlateCarree(),
                          color=colors_depths[i])
        
     # Add standard features
    ax.add_feature(cfeature.LAND, color='grey')
    ax.coastlines(lw=1, resolution='110m')
    ax.gridlines(draw_labels=False)
    ax.set_position([0.03, 0.05, 0.8, 0.9])

    # Add color bar
    axi = fig.add_axes([0.025, 0.05, 0.8, 0.02])  # Adjust the position and size
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    sm = plt.cm.ScalarMappable(cmap=blues_cm, norm=norm)
    cbar = fig.colorbar(mappable=sm,
                cax=axi,
                orientation='horizontal',  # Set the orientation to horizontal
                extend='min',
                ticks=depths,
                label='Depth (m)')
    cbar.ax.invert_xaxis()  # Reverse the direction of the color bar

    # Define an offset for moving the text downward
    text_offset = 0.3  # You can adjust this value as needed

    # Set font style with Times New Roman
    font1 = {'family': 'Times New Roman', 'weight': 'normal'}

    # Add labels for each point with adjusted latitude coordinates
    ax.text(ashes_lon, ashes_lat - text_offset, '{:.2f}, {:.2f}'.format(ashes_lon, ashes_lat),
            verticalalignment='top', horizontalalignment='center',
            transform=ccrs.PlateCarree(), fontsize=10, color='black', zorder=10, fontdict=font1)

    # Add text annotation for the Pacific Ocean
    pacific_lon = -130
    pacific_lat = 42.5

    # Set font style with Times New Roman
    font2 = {'family': 'Times New Roman', 'weight': 'normal', 'style': 'italic'}

    # You can also apply the same font style to other text elements in your plot
    ax.text(pacific_lon, pacific_lat, 'Pacific Ocean', transform=ccrs.PlateCarree(),
            fontsize=18, color='Grey', ha='center', fontdict=font2)

    # Set latitude and longitude ticks and labels
    ax.set_xticks(np.arange(-135, -119, 2), crs=ccrs.PlateCarree())
    ax.set_yticks(np.arange(40, 51, 2), crs=ccrs.PlateCarree())
    lon_formatter = cartopy.mpl.ticker.LongitudeFormatter()
    lat_formatter = cartopy.mpl.ticker.LatitudeFormatter()
    ax.xaxis.set_major_formatter(lon_formatter)
    ax.yaxis.set_major_formatter(lat_formatter)

    # Rotate x-axis tick labels
    plt.xticks(rotation=45)

```
